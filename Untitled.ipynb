```python
# âœ… Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# âœ… Step 2: Load CIFAR-10 Dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X = np.concatenate((X_train, X_test))
y = np.concatenate((y_train, y_test)).flatten()

# âœ… Step 3: Filter Cat (label 3) and Dog (label 5) images only
cat_dog_idx = np.where((y == 3) | (y == 5))[0]
X = X[cat_dog_idx]
y = y[cat_dog_idx]

# âœ… Step 4: Convert image to grayscale + flatten for ML model
X_gray = X.mean(axis=3)  # Convert to grayscale by averaging RGB
X_flat = X_gray.reshape(X_gray.shape[0], -1)

# âœ… Step 5: Change labels to binary: cat=0, dog=1
y_binary = np.where(y == 3, 0, 1)

# âœ… Step 6: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_flat, y_binary, test_size=0.2, random_state=42)

# âœ… Step 7: Train a simple SVM model
model = SVC(kernel='linear')
model.fit(X_train[:2000], y_train[:2000])




# âœ… Step 8: Predict & Evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"ðŸŽ¯ Model Accuracy: {accuracy*100:.2f}%")

# âœ… Step 9: Show 5 sample predictions
fig, axes = plt.subplots(1, 5, figsize=(15, 3))
for i, ax in enumerate(axes):
    ax.imshow(X_test[i].reshape(32, 32), cmap='gray')
    label = "Dog" if y_pred[i] == 1 else "Cat"
    ax.set_title(f"Pred: {label}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```
